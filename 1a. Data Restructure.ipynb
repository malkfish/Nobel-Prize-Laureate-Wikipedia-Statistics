```python
columns = ["YEAR", "LAUREATE", "NAME_ORDER", "URL", "FIELD"]
RESTRUCTURED_DF = pd.DataFrame(columns=columns)
NOBEL_PRIZE_FIELDS = ["Physics", "Chemistry", "Medicine", "Literature", "Peace", "Economy"]

all_rows = []  # Store all rows before creating the DataFrame

for field in NOBEL_PRIZE_FIELDS:
    print(f"Processing {field} laureates.")
    df = pd.read_csv(f"outputs\laureates_by_field\{field}.csv")

    for row in df.itertuples(index=False):
        year = int(row[0])
        laureates = row[1:4]  # All laureate name columns
        urls = row[4:7]  # Corresponding URLs
            
        for i, (laureate, url) in enumerate(zip(laureates, urls), start=1):
            if laureate != "No Record":
                all_rows.append([year, laureate, i, url, field])

# Create DataFrame in one go
RESTRUCTURED_DF = pd.DataFrame(all_rows, columns=columns)
        
```

```python
RESTRUCTURED_DF.loc[RESTRUCTURED_DF['LAUREATE'] == "No Record", 'NAME_ORDER'] = 0
RESTRUCTURED_DF = RESTRUCTURED_DF.sort_values(by=["YEAR", "FIELD", "NAME_ORDER",])
RESTRUCTURED_DF
```

```python
RESTRUCTURED_DF.to_csv("outputs/Nobel_Laureates_Clean.csv", index=False)
```
