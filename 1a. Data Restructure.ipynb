```python
columns = ["YEAR", "LAUREATE", "NAME_ORDER", "URL", "FIELD"]
RESTRUCTURED_DF = pd.DataFrame(columns=columns)
NOBEL_PRIZE_FIELDS = ["Physics", "Chemistry", "Medicine", "Literature", "Peace", "Economy"]

for field in NOBEL_PRIZE_FIELDS:
    print(f"Processing {field} laureates.")
    df = pd.read_csv(f"outputs\laureates_by_field\{field}.csv", index_col=0)

    if field == "Literature":
        for row in df.itertuples():
            year = int(row[1])
            df_row = [year, row[2], 1, row[4], field]
            RESTRUCTURED_DF.loc[-1] = df_row
            RESTRUCTURED_DF.index = RESTRUCTURED_DF.index + 1  # shifting index
            RESTRUCTURED_DF = RESTRUCTURED_DF.sort_index()  # sorting by index

            if row[3] != "No Record":
                df_row = [year, row[3], 2, row[5], field]
                RESTRUCTURED_DF.loc[-1] = df_row
                RESTRUCTURED_DF.index = RESTRUCTURED_DF.index + 1  # shifting index
                RESTRUCTURED_DF = RESTRUCTURED_DF.sort_index()  # sorting by index
    else:
        #insanity begins
        for row in df.itertuples():
            year = int(row[1])
            df_row = [year, row[2], 1, row[5], field]
            RESTRUCTURED_DF.loc[-1] = df_row
            RESTRUCTURED_DF.index = RESTRUCTURED_DF.index + 1  # shifting index
            RESTRUCTURED_DF = RESTRUCTURED_DF.sort_index()  # sorting by index

            if row[3] != "No Record":
                df_row = [year, row[3], 2, row[6], field]
                RESTRUCTURED_DF.loc[-1] = df_row
                RESTRUCTURED_DF.index = RESTRUCTURED_DF.index + 1  # shifting index
                RESTRUCTURED_DF = RESTRUCTURED_DF.sort_index()  # sorting by index

            if row[4] != "No Record":
                df_row = [year, row[4], 3, row[7], field]
                RESTRUCTURED_DF.loc[-1] = df_row
                RESTRUCTURED_DF.index = RESTRUCTURED_DF.index + 1  # shifting index
                RESTRUCTURED_DF = RESTRUCTURED_DF.sort_index()  # sorting by index
```
